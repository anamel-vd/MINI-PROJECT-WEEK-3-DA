```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/miriamg/Documents/GitHub_Ironhack/Week 3 SQL/Mini project/Clean_energy_countries.csv')
```

```python
df.dropna(inplace=True)
df.reset_index(inplace=True)
```

```python
df.rename(columns= {'Entity': 'Country', 'Renewable energy share in the total final energy consumption (%)': 're_share_in_total_ec_in_percentage'}, inplace= True)
df.columns
```

```python
df.drop(columns= ['index', 'Unnamed: 0', 'Electricity from fossil fuels (TWh)', 'Electricity from nuclear (TWh)', 'Energy intensity level of primary energy (MJ/$2017 PPP GDP)', 'CO2', 'Density', 'Land', 'Latitude',
       'Longitude'], inplace= True)

df
```

```python
df.rename(columns=lambda x: x.lower().replace("%", "").replace("(", "").replace(")", "").replace("/", "").replace("-", "").replace(" ", "_"), inplace=True)
df.columns
```

```python
df.rename(columns= {'access_to_electricity__of_population': 'population_access_to_electricity',
       'lowcarbon_electricity__electricity': 'lowcarbon_electricity'}, inplace= True)
df.columns
```

```python
countries = sorted(list(df['country'].unique()))
countries
```

```python
country_id = list(range(1,len(countries)+1))
country_id
```

## Create country table

```python
country_mapping = dict(zip(countries,country_id))
country_mapping
```

```python
df['country_id'] = df['country'].map(country_mapping)
```

```python
df['country_id']
```

```python
country_dict = {'country_id':country_id, 'country_name':countries}
country_df = pd.DataFrame(country_dict)
country_df.info()
```

```python
country_df.to_csv("countries_renewables.csv")
```

## Create the indicators table

```python
indicators_df = df[['year', 'population_access_to_electricity',
       'access_to_clean_fuels_for_cooking',
       're_share_in_total_ec_in_percentage', 'electricity_from_renewables_twh',
       'lowcarbon_electricity',
       'primary_energy_consumption_per_capita_kwhperson', 'gdp_growth',
       'gdp_per_capita', 'country_id']]

indicators_df = indicators_df.drop_duplicates()
```

```python
indicators_df.columns
```

## Create csv's

```python
country_df.to_csv('country_table.csv')
indicators_df.to_csv('indicators_table.csv')
```
